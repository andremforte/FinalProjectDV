```python
b = ['services_trade','merchandise_trade']
     
for col in b:
     df[col] = df[col].astype(str)

a['text'] = df['country'] + '<br>' + \
    'Trade in Services = ' + df['services_trade'] + ' % of GDP' + '<br>' + \
    'Merchandise trade = ' + df['merchandise_trade'] + '% of GDP' + '<br>'


data_choropleth = dict(type='choropleth',
                       locations=a['country_code'],  #There are three ways to 'merge' your data with the data pre embedded in the map
                       z=a['GDP'],
                       text=a['text'],
                       colorscale='Blues',
                       colorbar_title="$US"
                      )

layout_choropleth = dict(geo=dict(scope='europe',  #default
                                  projection=dict(type='orthographic'
                                                 ),
                                  #showland=True,   # default = True
                                  landcolor='black',
                                  lakecolor='white',
                                  showocean=True,  
                                # default = False
                                  oceancolor='azure'
                                 ),
                         
                         title=dict(text='2018 European GDP ($ US) Choropleth Map',
                                    x=.5 # Title relative position according to the xaxis, range (0,1)
                                   )
                        )
```

```python
fig_choropleth = go.Figure(data=data_choropleth, layout=layout_choropleth)

fig_choropleth.show(renderer = 'browser')
```
